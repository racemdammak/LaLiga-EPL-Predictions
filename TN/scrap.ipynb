```python
import pandas as pd
import json
```

```python
# Open the file and load its contents
with open("tunisian_league_data.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

```python
standings = list()

for i in range(len(data["standings"][0]["rows"])):
    team = data["standings"][0]["rows"][i]["team"]["name"]
    points = data["standings"][0]["rows"][i]["points"]
    wins = data["standings"][0]["rows"][i]["wins"]
    draws = data["standings"][0]["rows"][i]["draws"]
    losses = data["standings"][0]["rows"][i]["losses"]
    goals_for = data["standings"][0]["rows"][i]["scoresFor"]
    goals_against = data["standings"][0]["rows"][i]["scoresAgainst"]

    standings.append({
        "Team": team,
        "Pts": points,
        "Wins": wins,
        "Draws": draws,
        "Losses": losses,
        "GoalsFor": goals_for,
        "GoalsAgainst": goals_against,
        "GoalsDifference": goals_for - goals_against,
        "Rank": i + 1
    })
```

```python
data = pd.DataFrame(standings)
data
```

```python
data.to_csv("data/tunisian_league_standings.csv", index=False, encoding="utf-8")
```
