```python
import pandas as pd
import json
```

```python
# Open the file and load its contents
with open("data/tunisian_league_data.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

```python
standings = list()

for i in range(len(data["standings"][3]["rows"])):
    team = data["standings"][3]["rows"][i]["team"]["name"]
    position = data["standings"][3]["rows"][i]["position"]
    points = data["standings"][3]["rows"][i]["points"]
    wins = data["standings"][3]["rows"][i]["wins"]
    draws = data["standings"][3]["rows"][i]["draws"]
    losses = data["standings"][3]["rows"][i]["losses"]
    goals_for = data["standings"][3]["rows"][i]["scoresFor"]
    goals_against = data["standings"][3]["rows"][i]["scoresAgainst"]

    standings.append({
        "Team": team,
        "Position": position,
        "Pts": points,
        "Wins": wins,
        "Draws": draws,
        "Losses": losses,
        "GoalsFor": goals_for,
        "GoalsAgainst": goals_against,
        "GoalsDifference": goals_for - goals_against,
    })
```

```python
all = pd.DataFrame(standings)
all
```

```python
all.to_csv("data/2022/group2/group2_standings2022.csv", index=False)
```
