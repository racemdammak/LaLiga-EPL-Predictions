```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://fbref.com/en/comps/12/La-Liga-Stats"
    
# Add headers to mimic a browser request
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8'
    }

response = requests.get(url, headers=headers)
response.raise_for_status()
print(response)
```

```python
soup = BeautifulSoup(response.content, "html.parser")

results_table = soup.find('table', {'id': 'results2025-2026121_overall'})
df = pd.read_html(str(results_table))[0]


columns_to_drop = ['Attendance', 'Notes', 'Top Team Scorer', 'Attendance', 'Goalkeeper', 'W', 'L', 'D', 'GF', 'GA', 'GD', 'MP']
df.drop(columns=columns_to_drop, inplace=True)

df
```

```python
df.to_csv('laliga26-table.csv', index=False)
```
